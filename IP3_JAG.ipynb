```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
car_sales=pd.read_csv("car-sales.csv")
```

```python
car_sales.head()
```

```python
plt.plot(car_sales['Price'])
```

```python
plt.plot(car_sales['Odometer (KM)'])
```

```python
car_sales.info()
```

```python
car_sales['Price']=car_sales['Price'].str.replace("[\$\,]","")
```

```python
car_sales.head()
```

```python
car_sales['Price']=car_sales['Price'].astype('int64')
car_sales.info()
```

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

```python
car_sales['Sales Date']=pd.date_range("2/22/2022", periods=len(car_sales))
car_sales.head()
```

```python
plt.plot(car_sales['Price'], car_sales['Total Sales'])
plt.title('Price vs Total Sales Line Plot')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title('Odometer vs Price Scatter Plot')
plt.xlabel('Odometer')
plt.ylabel('Price')
```

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.ylabel('Odometer')
plt.xlabel('Make')
plt.title('Make vs Odometer Bar Plot')
plt.show()
```

```python
plt.hist(car_sales['Make'])
plt.title('Histogram of car make')
plt.xlabel('Make')
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize = (12,10))

ax1.plot(car_sales['Price'], car_sales['Total Sales'])
ax1.set_title('Price vs Total Sales Line Plot')
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
ax2.set_title('Odometer vs Price Scatter Plot')
ax2.set_xlabel('Odometer')
ax2.set_ylabel('Price')

ax3.bar(car_sales['Make'], car_sales['Odometer (KM)'])
ax3.set_ylabel('Odometer')
ax3.set_xlabel('Make')
ax3.set_title('Make vs Odometer Bar Plot')

ax4.hist(car_sales['Make'])
ax4.set_title('Histogram of car make')
ax2.set_xlabel('Make')

plt.show()
```
